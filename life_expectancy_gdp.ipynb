```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('all_data.csv')
```

```python
print(data.head())
```

```python
print(data.tail())
```

```python
print(data.info())
```

```python
sns.scatterplot(x = 'Country', y = 'Life expectancy at birth (years)', hue = 'Year', data = data)
plt.show()
plt.clf()
```

```python
data_no_zimbabwe = data[data['Life expectancy at birth (years)'] > 65]
sns.scatterplot(y = 'Life expectancy at birth (years)', x = 'GDP', hue = 'Country', data = data_no_zimbabwe)
```

```python
zimbabwe_data = data[data.Country == 'Zimbabwe']
#sns.scatterplot(x = 'Life expectancy at birth (years)', y = 'GDP', data = zimbabwe_data)
sns.lmplot(x = 'Life expectancy at birth (years)', y = 'GDP', hue = 'Year', data = zimbabwe_data, line_kws = {'color': 'purple'})
plt.title('Life Expectancy vs GDP for Zimbabwe')
plt.show()
plt.clf()
```

```python
sns.lmplot(x = 'Life expectancy at birth (years)', y = 'GDP', data = data[data.Country == 'Chile'], line_kws = {'color': 'red'})
plt.title('Life Expentancy vs GDP for Chile')
plt.show()
plt.clf()

sns.lmplot(x = 'Life expectancy at birth (years)', y = 'GDP', data = data[data.Country == 'China'], line_kws = {'color': 'red'})
plt.title('Life Expentancy vs GDP for China')
plt.show()
plt.clf()

sns.lmplot(x = 'Life expectancy at birth (years)', y = 'GDP', data = data[data.Country == 'Germany'], line_kws = {'color': 'red'})
plt.title('Life Expentancy vs GDP for Germany')
plt.show()
plt.clf()
```

```python
sns.boxplot(x = 'Country', y = 'Life expectancy at birth (years)', data = data)
plt.show()
plt.clf()
```

```python
data_Zimbabwe = data[data.Country == 'Zimbabwe']
sns.scatterplot(y=data_Zimbabwe['Life expectancy at birth (years)'], x = data_Zimbabwe['Year'])
plt.title('Life expectancy in Zimbabwe with time')
plt.show()
plt.clf()
```
